## 7.2.1 删除重复值

          目的 :  删除重复值
          工具 : duplicated()   drop_duplicates() drop_duplicates([value], keep = 'last) 

```python
data = pd.DataFrame({'a':['one','two'] * 3 + ['two'], 'b':[1,1,2,3,3,4,4] })
```

```python
data
```

```python
data.duplicated()
```

```python
data.drop_duplicates()
```

```python
data
```

```python
data['new'] = range(7)
```

```python
data
```

```python
data.drop_duplicates(['a'])
```

```python
data
```

```python
data.drop_duplicates(['a','b'],keep='last')
```

```python
data.drop_duplicates(['a','b'])
```
