       使用需求总结
       1  生成 整数 np.random.randint(low,high,size)
       2 生成 线性数组  np.linspace(start,end,num) num: 总数
       3 生成正态分布数据 np.random.normal(loc,scale,size) loc :中心 scale :标准差 大矮胖 小瘦高  size: 默认 none 只输出一个值_
       4 从一个均匀分布中随机采样 左闭右开 
       np.random.uniform(low,high,size) 
       low: 采样下界 type:float def:0
       high:采样上界 type:float def 1
       size: 样本数目  type: int 或 tuple
