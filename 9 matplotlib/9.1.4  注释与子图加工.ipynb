```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(100).cumsum())
```

```python
fig  = plt.figure()
ax = fig.add_subplot(1,1,1)
rect = plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3)
ax.add_patch(rect)
circle = plt.Circle((0.7,0.2),0.15,color='c',alpha=0.5)
ax.add_patch(circle)
polygon = plt.Polygon([[0.15, 0.15], [0.35, 0.4] ,[0.2,0.6],[0.32,0.1]],color='m',alpha=0.6)
ax.add_patch(polygon)
```
