    9.2.1折线图

```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series( np.random.randn(10).cumsum(), index = np.arange(0,10) )
```

```python
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10,4).cumsum(0),
                 columns = ['A','B','C','D'],
                  index = np.arange(0,100,10)
                 )
df.plot()
```
