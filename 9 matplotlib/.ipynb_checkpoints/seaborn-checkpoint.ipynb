```python
# %matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
#np.random.seed(sum(map(ord, "aesthetics")))
```

```python
np.random.seed(sum(map(ord,"aesthetics")))
```

```python
def sinplot (flip = 1):
    x = np.linspace(1,14,100)
    for i in range(1,6):
        plt.plot(x, np.sin(x-1.2 + i * .5 ) *(7-i) * flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

     控制风格 : axes_style(), set_style()
     缩放绘图 : plotting_context(), set_content()

     Seaborn 五种绘图风格 (darkgrid whitegrid dark white ticks)

```python
sns.set_style("darkgrid")
data = np.random.normal(size=(20,6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20,6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style("ticks")
data = np.random.normal(size=(20,6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style("white")
data = np.random.normal(size=(20,6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style("dark")
data = np.random.normal(size=(20,6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

   set_style :  背景颜色 white dark  是否有横线格 and ticks 

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20,4)) + np.arange(4) / 2
sns.boxplot(data=data)
```
